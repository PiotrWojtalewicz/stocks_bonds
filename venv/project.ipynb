```python
### libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import yfinance as yf
from datetime import datetime as dt
from datetime import timedelta
```

```python
#set the time 
endtime = dt.now()
starttime = endtime-timedelta(days=1000)
```

```python
#import necessery data from yahoo finance
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'IEF']
date = yf.download(tickers, starttime, endtime)
close_prices = date['Adj Close']
```

```python
dividends = {}
for ticker in tickers:
    stock = yf.Ticker(ticker)
    dividends[ticker] = stock.dividends
    
dividends = pd.DataFrame(dividends)
```

```python
dividends.tail()
```

```python
#daily returns
daily_returns = close_prices.pct_change()
daily_returns
```

```python
#monthly returns
monthly_returns = close_prices.resample('ME').ffill().pct_change()
monthly_returns
```

```python
#annual returs
annual_returns = close_prices.resample('YE').ffill().pct_change()
annual_returns
```
